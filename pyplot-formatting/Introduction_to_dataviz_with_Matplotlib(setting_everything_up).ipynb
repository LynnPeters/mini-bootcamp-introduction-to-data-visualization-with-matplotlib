We can make multiple graphics in one figure. This goes very well for comparing charts or sharing data from several types of charts easily with a single image.
The .subplot() method adds multiple plots in one figure. It takes three arguments:
nrows: number of rows in the figure
ncols: number of columns in the figure
index: index of the plot
Let's see how the figures plot variables with different row and column configurations.

```python
#multiple plots in one figure 
fig, ax  = plt.subplots()
print(fig,ax)
```

```python
#1 row, 2 olumns-method 1 
plt.subplot(1,2,1)#subplot of the row,columns,index
plt.plot(x.work_force, x.income, 'go')
plt.title("Income vs Work Force")

plt.subplot(1,2,2)#index2 means the plot will be shown second
plt.plot(x.hospital_beds, x.income, "^r")
plt.title("Income vs Hospital Beds")

plt.suptitle("Sub Plots")#adds a title to the middle of the plots
```

```python
#1 row and 2 columns method 2 
fig, (ax1, ax2)= plt.subplots(1,2)#roww, columns
fig.suptitle('Horizontally stacked subplots')
ax1.plot(x.work_force, x.income, 'go')
ax1.set_title("Income vs Workforce")

ax2.plot(x.hospital_beds, x.income, '^r')
ax2.set_title("Income vs Hospital Beds")
```

```python
#overlapping values(where values are shared on a specific axis)
plt.subplot(2,1,1)#shows nr of rows, nr of columns, nr in index
plt.plot(x.work_force, x.income, 'go')

plt.subplot(2,1,2)#rows,columns,index
plt.plot(x.hospital_beds, x.income, "r^")

plt.suptitle("SubPlots")

#a two row and one column graph 
```

Think of this as a 22 grid. You can access the different positions of the axis and plot your graphs there as follows:

```python
fig, ax  = plt.subplots(nrows=2, ncols =2, figsize=(6,6))#grid of 2 rows, 2 columns and 6x6 fig size 
ax[0,0].plot(x.work_force, x.income, "go")#think of nrs as coord in a graph (0,0)=liesin the upper left axis 
ax[0,1].plot(x.work_force, x.income, "bo")#upper right axis 
ax[1,0].plot(x.work_force, x.income, "yo")#lower left axis 
ax[1,1].plot(x.work_force, x.income, "ro")#lower right 
```
